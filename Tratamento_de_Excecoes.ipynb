Exercício 1: Validação de Nome
Este script verifica se o valor digitado é uma string válida (não permitindo números isolados).

```python
# Exercício 1: Validar se o nome é uma string válida
def validar_nome():
    try:
        nome = input("Digite um nome: ").strip()

        # Verificando se a string contém apenas letras e espaços
        if not nome.replace(" ", "").isalpha() or nome == "":
            raise ValueError("O nome deve conter apenas letras e não pode estar vazio.")

        print(f"Nome válido: {nome}")

    except ValueError as e:
        print(f"Erro: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

validar_nome()
```

Exercício 2: Divisão com Tratamento de Erros.
Trata divisões por zero e entradas que não são numéricas.

```python
def divisao_segura():
    try:
        num1 = int(input("Digite o primeiro número (numerador): "))
        num2 = int(input("Digite o segundo número (denominador): "))

        resultado = num1 / num2

    except ZeroDivisionError:
        print("Erro: Não é possível realizar divisão por zero.")
    except ValueError:
        print("Erro: Um dos valores digitados não é um número válido.")
    else:
        print(f"Resultado da divisão: {resultado}")

divisao_segura()
```

Exercício 3: Fatorial em Lista
Calcula o fatorial tratando elementos não numéricos e números negativos.

```python
import math

# Exercício 3: Fatorial de uma lista de números
def calcular_fatoriais(lista):
    print(f"Analisando a lista: {lista}")
    for item in lista:
        try:
            # Verifica se é número e se é negativo
            if not isinstance(item, int):
                raise TypeError(f"O elemento '{item}' não é um número inteiro.")
            if item < 0:
                raise ValueError(f"O número {item} é negativo. Não existe fatorial de número negativo.")

            print(f"Fatorial de {item}: {math.factorial(item)}")

        except (TypeError, ValueError) as e:
            print(f"Erro no elemento: {e}")

# Teste com lista variada
minha_lista = [5, "texto", 3, -1, 0]
calcular_fatoriais(minha_lista)
```

Exercício 4: Contador de Palavras e Linhas
Lê um arquivo e conta seus elementos, verificando se o arquivo existe no diretório.

```python
# Exercício 4: Contar palavras e linhas de um arquivo
def analisar_arquivo(nome_arquivo):
    try:
        with open(nome_arquivo, "r", encoding="utf-8") as f:
            linhas = f.readlines()
            num_linhas = len(linhas)

            # Conta palavras separando por espaços em cada linha
            total_palavras = 0
            for linha in linhas:
                total_palavras += len(linha.split())

            print(f"Análise do arquivo '{nome_arquivo}':")
            print(f"- Total de linhas: {num_linhas}")
            print(f"- Total de palavras: {total_palavras}")

    except FileNotFoundError:
        print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado no diretório.")
    except Exception as e:
        print(f"Erro inesperado: {e}")

# Criando um arquivo de teste rápido para o Colab
with open("exemplo.txt", "w") as teste:
    teste.write("Olá Mundo\nEste é um exercício de Python\nFim do arquivo.")

analisar_arquivo("exemplo.txt")
```

Exercício 5: Formatador de Nomes e Armazenamento
Converte o nome para o formato de título (Capitalize) e salva em arquivo.

```python
# Exercício 5: Formatar nome e salvar em arquivo
def formatar_e_salvar():
    try:
        nome_completo = input("Digite seu nome completo: ").strip()

        # Validação simples de nome
        if nome_completo == "" or any(char.isdigit() for char in nome_completo):
            raise ValueError("Nome inválido! Certifique-se de não usar números ou deixar vazio.")

        # Converte primeira letra de cada palavra em Maiúscula
        nome_formatado = nome_completo.title()

        nome_arquivo = "nomes_formatados.txt"

        # Salva no arquivo (modo 'a' para adicionar sem apagar o anterior)
        with open(nome_arquivo, "a", encoding="utf-8") as f:
            f.write(nome_formatado + "\n")

        print(f"Sucesso! '{nome_formatado}' foi armazenado em {nome_arquivo}.")

    except ValueError as e:
        print(f"Erro de validação: {e}")
    except PermissionError:
        print("Erro: Sem permissão para escrever no diretório.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

formatar_e_salvar()
```
